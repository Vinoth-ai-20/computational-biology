## Step 1 — Motivation

This module exists because of one uncomfortable fact: you cannot do computational
biology without biology. A data scientist who doesn't understand the data is not
doing science — they're running pipelines. Every downstream module (genomics,
transcriptomics, ML for biology, structural biology) requires the vocabulary,
mental models, and intuitions built here.

This notebook does not teach biology. It maps the gap — what a computational
biologist needs to know that a pure computer scientist does not — and sets up
a 12-month contract for filling it.

---
## Step 2 — Intuition: What 'Literate' Means Here

**A bench biologist** needs to run gel electrophoresis, handle cells under a hood,
and calibrate a pipette. You don't need any of that.

**A computational biologist** needs to:
- Read a genomics paper and understand what the experiment measured
- Know why a particular normalization is applied to expression data
- Explain in an interview what RNA-seq measures and why it matters
- Write code that operates on biological sequences without confusing a codon and a kmer
- Recognize when a result is biologically implausible, even if the statistics look fine

That is the bar this module builds to — not PhD-level molecular biology,
but enough to reason correctly about biological data.
