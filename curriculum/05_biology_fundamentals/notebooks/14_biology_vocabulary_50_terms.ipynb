## The 50 Terms

### Group A — Molecular Biology (NB03–NB04)

| # | Term | Definition |
|---|------|------------|
| 1 | Central dogma | DNA → RNA → Protein; information flows in this direction under normal conditions |
| 2 | Transcription | The process of copying DNA to pre-mRNA; occurs in the nucleus |
| 3 | Translation | The process of decoding mRNA into a protein chain at the ribosome |
| 4 | Codon | A triplet of mRNA nucleotides that specifies one amino acid (or stop) |
| 5 | Exon | Portion of the pre-mRNA retained in the mature mRNA after splicing |
| 6 | Intron | Portion of the pre-mRNA spliced out; not present in mature mRNA |
| 7 | Open reading frame | A continuous stretch of codons starting at AUG and ending at a stop codon |
| 8 | Promoter | DNA region upstream of a gene where RNA polymerase binds to initiate transcription |
| 9 | Enhancer | Regulatory DNA element that can be far from a gene but loops to increase transcription |
| 10 | 5'UTR / 3'UTR | Untranslated regions flanking the coding sequence; involved in stability and regulation |

### Group B — Genetics (NB05–NB07)

| # | Term | Definition |
|---|------|------------|
| 11 | Allele | One version of the DNA sequence at a specific locus |
| 12 | Locus | A specific position on a chromosome |
| 13 | Heterozygous | Having two different alleles at a locus (Aa) |
| 14 | Homozygous | Having two identical alleles at a locus (AA or aa) |
| 15 | Dominant | Allele whose effect is observed even in heterozygous state |
| 16 | Recessive | Allele whose effect is only observed in homozygous state |
| 17 | Phenotype | Observable trait resulting from genotype + environment |
| 18 | Genotype | The alleles an organism carries at a given locus or across the genome |
| 19 | Hardy-Weinberg equilibrium | Expected genotype frequencies (p², 2pq, q²) under random mating with no disturbing forces |
| 20 | Genetic drift | Random change in allele frequency due to finite population size |

### Group C — Genomics and Variation (NB04, NB09, NB12)

| # | Term | Definition |
|---|------|------------|
| 21 | SNP | Single nucleotide polymorphism: common DNA variant at a single position (AF > 1%) |
| 22 | Indel | Insertion or deletion of one or more nucleotides |
| 23 | Copy number variant (CNV) | Duplication or deletion of a large genomic segment (>1 kb) |
| 24 | Loss of function (LoF) | Variant that disables the gene product (frameshift, nonsense, splice) |
| 25 | Gain of function | Variant that creates a new or enhanced activity in the protein |
| 26 | Missense variant | Point mutation that changes one amino acid to another |
| 27 | Synonymous variant | Point mutation that does not change the amino acid (same codon effect) |
| 28 | Allele frequency | Proportion of alleles in a population that are of a particular type |
| 29 | Linkage disequilibrium (LD) | Non-random association of alleles at two loci (alleles inherited together) |
| 30 | Haplotype | A set of alleles on the same chromosome that are inherited together |

### Group D — Evolution and Population Genetics (NB07–NB09)

| # | Term | Definition |
|---|------|------------|
| 31 | Natural selection | Differential survival and reproduction based on heritable traits |
| 32 | Selection coefficient (s) | Relative fitness advantage or disadvantage of an allele |
| 33 | Neutral theory | Most molecular variation is not subject to selection; fixed by drift |
| 34 | dN/dS | Ratio of non-synonymous to synonymous substitution rates; >1 indicates positive selection |
| 35 | Phylogenetic tree | Diagram showing evolutionary relationships between taxa or sequences |
| 36 | Clade | A group of organisms sharing a common ancestor; a monophyletic group |
| 37 | Molecular clock | Hypothesis that substitutions accumulate at a roughly constant rate over time |
| 38 | Founder effect | Loss of genetic variation when a new population is established by a small number |
| 39 | Selective sweep | Reduction of diversity near a beneficial allele as it rises to fixation |
| 40 | Effective population size (Ne) | Size of an ideal population with the same drift rate as the real population |

### Group E — Cell Biology and Regulation (NB02, NB10)

| # | Term | Definition |
|---|------|------------|
| 41 | Transcription factor | Protein that binds specific DNA sequences to activate or repress transcription |
| 42 | Chromatin | DNA wrapped around histone proteins; compaction state controls gene accessibility |
| 43 | Euchromatin | Open, transcriptionally active chromatin |
| 44 | Heterochromatin | Closed, transcriptionally silent chromatin |
| 45 | Epigenetics | Heritable changes in gene expression not caused by changes in DNA sequence |
| 46 | CpG island | GC-rich region near many gene promoters; methylation state affects expression |
| 47 | Phosphorylation | Addition of a phosphate group to a protein; often activates or deactivates it |
| 48 | Apoptosis | Programmed cell death; dysregulated in cancer |
| 49 | Cell cycle | Ordered sequence of events (G1→S→G2→M) by which a cell grows and divides |
| 50 | Oncogene / tumour suppressor | Cancer gene that promotes growth (GoF) or inhibits growth (LoF when mutated) |

```python
import numpy as np
import matplotlib.pyplot as plt

terms = [
    # Group A — Molecular Biology
    'Central dogma', 'Transcription', 'Translation', 'Codon', 'Exon',
    'Intron', 'Open reading frame', 'Promoter', 'Enhancer', '5\'/3\'UTR',
    # Group B — Genetics
    'Allele', 'Locus', 'Heterozygous', 'Homozygous', 'Dominant',
    'Recessive', 'Phenotype', 'Genotype', 'Hardy-Weinberg', 'Genetic drift',
    # Group C — Genomics and Variation
    'SNP', 'Indel', 'CNV', 'Loss of function', 'Gain of function',
    'Missense', 'Synonymous', 'Allele frequency', 'LD', 'Haplotype',
    # Group D — Evolution
    'Natural selection', 'Selection coeff.', 'Neutral theory', 'dN/dS', 'Phylogenetic tree',
    'Clade', 'Molecular clock', 'Founder effect', 'Selective sweep', 'Ne',
    # Group E — Cell Biology
    'Transcription factor', 'Chromatin', 'Euchromatin', 'Heterochromatin', 'Epigenetics',
    'CpG island', 'Phosphorylation', 'Apoptosis', 'Cell cycle', 'Oncogene/TSG',
]

# FILL IN YOUR SCORES (0=unknown, 1=vague, 2=clear, 3=can teach)
scores = [0] * 50  # EDIT THESE AFTER GOING THROUGH THE TABLE

groups = ['Molecular Biology', 'Genetics', 'Genomics & Variation', 'Evolution', 'Cell Biology']
group_scores = [scores[i*10:(i+1)*10] for i in range(5)]

print(f"Total score: {sum(scores)}/150  ({sum(scores)/150:.0%})")
print("\nBy group:")
for g, gs in zip(groups, group_scores):
    print(f"  {g:<25}: {sum(gs)}/30  ({sum(gs)/30:.0%})")

print("\nTerms to review (score < 2):")
for i, (term, score) in enumerate(zip(terms, scores)):
    if score < 2:
        print(f"  #{i+1:>2}: {term}  (score={score})")
```

```python
# Cell 6.2 — Radar chart of vocabulary mastery by group
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))

n_groups = 5
angles = np.linspace(0, 2 * np.pi, n_groups, endpoint=False).tolist()
angles += angles[:1]  # close the polygon

group_mean = [np.mean(gs) for gs in group_scores]
group_mean += group_mean[:1]

ax.plot(angles, group_mean, 'o-', color='steelblue', lw=2)
ax.fill(angles, group_mean, alpha=0.25, color='steelblue')

# Reference lines
for level in [1, 2, 3]:
    ax.plot(angles, [level] * (n_groups + 1), '--', color='gray', lw=0.5, alpha=0.5)

ax.set_thetagrids(np.degrees(angles[:-1]), groups, fontsize=9)
ax.set_ylim(0, 3)
ax.set_yticks([1, 2, 3])
ax.set_yticklabels(['Vague', 'Clear', 'Teach'], fontsize=7)
ax.set_title('Vocabulary Mastery by Module Area\n(0=unknown → 3=can teach)', y=1.1)

plt.tight_layout()
plt.show()
```
