## Step 6 — Python Implementation

````python
# Cell 6.1 — Verify Python version
import sys
print(f"Python {sys.version}")
assert sys.version_info >= (3, 11), f"Need Python ≥ 3.11, got {sys.version_info}"
````

````python
# Cell 6.2 — Verify all required packages
import importlib

REQUIRED = {
    "numpy":      ("numpy",     "1.26"),
    "scipy":      ("scipy",     "1.12"),
    "pandas":     ("pandas",    "2.2"),
    "matplotlib": ("matplotlib","3.8"),
    "seaborn":    ("seaborn",   "0.13"),
    "Bio":        ("biopython", "1.83"),
    "sklearn":    ("scikit-learn","1.4"),
    "jupytext":   ("jupytext",  "1.16"),
    "pytest":     ("pytest",    "8.0"),
    "ruff":       ("ruff",      "0.4"),
}

print(f"{'Package':<20} {'Version':<15} {'Min':<10} Status")
print("-" * 55)
all_ok = True
for import_name, (pkg_name, min_ver) in REQUIRED.items():
    try:
        mod = importlib.import_module(import_name)
        ver = getattr(mod, "__version__", "?")
        print(f"  {pkg_name:<18} {ver:<15} {min_ver:<10} ✓")
    except ImportError:
        print(f"  {pkg_name:<18} {'MISSING':<15} {min_ver:<10} ✗")
        all_ok = False

print()
print("All present." if all_ok else "Install missing packages from requirements.txt")
````

````python
# Cell 6.3 — Run scripts/verify_environment.py
import subprocess, pathlib

repo_root = pathlib.Path.cwd()
while repo_root != repo_root.parent:
    if (repo_root / ".git").exists(): break
    repo_root = repo_root.parent

verify_script = repo_root / "scripts" / "verify_environment.py"
if verify_script.exists():
    result = subprocess.run(["python", str(verify_script)], capture_output=True, text=True)
    print(result.stdout)
    if result.stderr: print("STDERR:", result.stderr)
else:
    print(f"Script not found: {verify_script}")
    print("Create scripts/verify_environment.py (covered in Module 00).")
````

````python
# Cell 6.4 — Verify compbio_utils is installed and importable
try:
    import compbio_utils
    print(f"compbio_utils version: {compbio_utils.__version__}")
    print("Submodules available:")
    for mod in ["sequence", "stats", "plotting", "io"]:
        try:
            importlib.import_module(f"compbio_utils.{mod}")
            print(f"  ✓  compbio_utils.{mod}")
        except ImportError:
            print(f"  ✗  compbio_utils.{mod} — missing")
except ImportError:
    print("compbio_utils not installed. Run: pip install -e utilities/compbio_utils")
````

````python
# Cell 6.5 — Implement compbio_utils.version() helper
# This is the Exercise: add version() to compbio_utils.
# For now, show what it should do:

def version_template():
    """
    Function to add to utilities/compbio_utils/__init__.py
    after completing the exercise.
    """
    # Read version from pyproject.toml
    import tomllib, pathlib
    pyproject = pathlib.Path(__file__).parent / "pyproject.toml"
    with open(pyproject, "rb") as f:
        data = tomllib.load(f)
    return data["project"]["version"]

# Show the expected semantic version format
import re
SEMVER = re.compile(r"^\d+\.\d+\.\d+(-[a-zA-Z0-9]+)?$")
test_versions = ["0.0.1-dev", "0.1.0", "1.0.0", "invalid", "1.2"]
print("Semantic version validation:")
for v in test_versions:
    match = "✓ valid" if SEMVER.match(v) else "✗ invalid"
    print(f"  {v:<15}  {match}")
````
