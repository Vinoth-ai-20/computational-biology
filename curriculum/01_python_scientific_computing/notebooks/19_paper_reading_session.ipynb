## Step 2 — Intuition

The Three-Pass Method (from Module 00 NB09) structures this session:

- **Pass 1 (5–10 min):** Title, abstract, section headings, figures, conclusion.
  Write 3 sentences: what is the paper, what is the main claim, what is the evidence.
- **Pass 2 (45–90 min):** Read in order. Note each design choice and its justification.
  Mark what is still unclear before asking.
- **Pass 3 (1–2 hrs, paper closed):** Reconstruct the core argument from memory.
  Record whether you succeeded unaided — this is the actual learning metric.

## Step 3 — Biological Background

**Why Bioconductor matters:**  
Bioconductor (2004) established R as the primary language for bioinformatics data
analysis. It defined the standard for open-source biological software: common data
structures (`ExpressionSet`), reproducible vignettes, peer-reviewed packages.

The Python equivalent — a fragmented landscape of PyPI packages — does not yet have
a comparable centralized ecosystem. Understanding *why* Bioconductor succeeded is
directly relevant to designing `compbio_utils` well.

````python
# Cell 6.2 — Timer for Pass 1 (run before reading)
import time

def timed_pass(pass_name: str, duration_minutes: int) -> None:
    print(f"\n{'='*50}")
    print(f"  {pass_name}")
    print(f"  Target: {duration_minutes} minutes")
    print(f"  Start time: {time.strftime('%H:%M:%S')}")
    print(f"  Stop time:  {time.strftime('%H:%M:%S', time.localtime(time.time() + duration_minutes*60))}")
    print(f"{'='*50}")
    print(f"  → Open the paper now and start reading.")
    print(f"  → Come back and fill in the section above when done.")

timed_pass("Pass 1 — Skim", duration_minutes=10)
````

````python
# Cell 6.4 — Pass 2 timer and guided questions
timed_pass("Pass 2 — Intent", duration_minutes=60)

print("\nGuided questions for Pass 2:")
questions = [
    "What problem was Bioconductor solving that CRAN/PyPI didn't?",
    "What is the ExpressionSet and why did a common data structure matter?",
    "How does Bioconductor ensure reproducibility? What mechanisms are described?",
    "What is the review process for Bioconductor packages?",
    "What features of this design could/should be adopted for compbio_utils?",
]
for i, q in enumerate(questions, 1):
    print(f"  {i}. {q}")
````

````python
# Cell 6.5 — Pass 3: reconstruction self-assessment
timed_pass("Pass 3 — Reconstruction (paper CLOSED)", duration_minutes=90)

print("\nAfter reconstructing, open the paper and check these specific claims:")
checks = [
    "Year Bioconductor was founded and initial number of packages",
    "The five design principles stated in the paper",
    "What ExpressionSet contains (the exact fields)",
    "How vignettes differ from documentation",
    "The role of version-coupled repositories",
]
for c in checks:
    print(f"  ☐ {c}")
````

## Step 7 — Visualization

````python
# Cell 7.1 — Three-pass time tracking visualization
import matplotlib.pyplot as plt

# Fill in actual times after completing the passes
passes = ["Pass 1\n(Skim)", "Pass 2\n(Intent)", "Pass 3\n(Reconstruct)"]
target_mins = [10, 60, 90]
actual_mins = [0,   0,  0]   # ← fill in actual minutes

x = range(len(passes))
fig, ax = plt.subplots(figsize=(7, 3))
ax.bar([i - 0.2 for i in x], target_mins, 0.4, label="Target", color="lightgray")
ax.bar([i + 0.2 for i in x], actual_mins, 0.4, label="Actual",
       color=["seagreen" if a > 0 else "tomato" for a in actual_mins], alpha=0.8)
ax.set_xticks(x)
ax.set_xticklabels(passes)
ax.set_ylabel("Minutes")
ax.set_title("Three-Pass reading time — Gentleman et al. 2004")
ax.legend()
plt.tight_layout()
plt.show()
````
