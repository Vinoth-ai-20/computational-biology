## Step 1 — Motivation

Module 05 gave you vocabulary and intuitions: what a gene is, what HWE says, what drift
does. This module goes one layer deeper — into the *models* and *methods* that researchers
actually use when they analyse population genomics data, infer evolutionary history, or
study the dynamics of cooperation. The gap between 'knows what HWE is' and 'can interpret
a GWAS population stratification correction' is what this module closes.

---
## Step 2 — What Module 05 Covered vs. What This Module Adds

| Topic | Module 05 level | Module 06 level |
|-------|-----------------|-----------------|
| Mendelian genetics | Punnett square, dominant/recessive | Linkage, recombination, epistasis, locus heterogeneity |
| Hardy-Weinberg | Definition, chi-square test | HWE violation over time, population structure effects |
| Genetic drift | Wright-Fisher intuition | Fixation probability formula, sweep signatures |
| Phylogenetics | UPGMA from scratch | Neighbour-joining, parsimony, likelihood concepts, molecular clock |
| Selection | Selection coefficient, Δp | Selection + drift interaction, ESS, replicator dynamics |
| Mutation | Frameshift and SNP types | Molecular mechanisms, repair pathways, rate estimation |
| New topics | — | Coalescent theory, evolutionary game theory, convergent evolution |

**The key conceptual advance from Module 05 to Module 06:**  
Module 05 asks 'what are the rules?' Module 06 asks 'what do the rules predict, and
how do we infer parameters from data?'

---
## Step 3 — The Research Literature Connection

**Where these concepts appear in current genomics papers:**

1. **Population stratification correction in GWAS** — requires understanding linkage
   disequilibrium (covered in NB02), population structure (Fst, covered in NB04),
   and the coalescent as a model for shared ancestry (NB05).

2. **Ancient DNA papers** — use the coalescent to estimate divergence times and population
   sizes from sequence diversity; molecular clock calibration (NB08) is central.

3. **Phylogenomics** — large-scale tree inference using ML (NB07) is now standard;
   understanding why NJ and UPGMA can fail (the Felsenstein zone) is needed to read
   methods sections critically.

4. **Evolutionary ecology papers** — evolutionary game theory (NB09) appears in
   cooperation, altruism, and ALife literature; relevant if Track B points to
   computational ecology or complex systems.

5. **Cancer genomics** — somatic mutation rates, convergent evolution of resistance
   mutations (NB10), and selection within tumours use exactly the population genetics
   framework developed here.
