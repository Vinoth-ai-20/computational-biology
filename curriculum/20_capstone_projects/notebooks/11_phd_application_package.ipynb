```python
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass, field
from typing import List, Optional
import textwrap

# ---- Personal statement structure ----

PS_TEMPLATE = """
PERSONAL STATEMENT TEMPLATE — European PhD (1,000 words max)

Paragraph 1 — Hook (50–80 words)
  What is the biological question that motivates everything?
  Why does it matter computationally? State this WITHOUT jargon.
  Do NOT start with "I have always been fascinated by..."
  Instead: open with a biological puzzle or a striking quantitative fact.

Paragraph 2 — Computational training (100–150 words)
  What can you actually DO? Name specific methods (NB, DESeq2-style normalization,
  MCMC, CNNs for sequences) + specific outcomes. Quantify wherever possible.
  (AUROC 0.983, 95.7% sensitivity, 7.3 pp improvement).
  DO NOT list modules or course names — demonstrate through capabilities.

Paragraph 3 — Biological depth (100–150 words)
  Show you understand the biology, not just the algorithms.
  Specific example: how does your RNA-seq CP01 connect to real gene regulation?
  What is the biological interpretation of differential expression?
  What does an epidemic model assume about population structure?

Paragraph 4 — Proposed research (200–250 words)
  State 1 specific question you would pursue with the supervisor.
  Describe 2–3 computational approaches (not a textbook list — YOUR choices).
  Acknowledge one major challenge and how you would approach it.
  Reference ONE specific paper from the supervisor's lab by name and year.

Paragraph 5 — Why THIS programme / supervisor (100–130 words)
  Demonstrate you have read their actual recent work (≥ 2 papers, ≥ 2018).
  Explain specifically how your existing skills complement their group's needs.
  Name one other group at the institution you would want to collaborate with.

Paragraph 6 — Closing (50–80 words)
  What do you plan to do AFTER the PhD? (Not "more research" — be specific.)
  One sentence summary of what you bring that is unusual or hard to replicate.
"""

print(PS_TEMPLATE)
```

```python
# ---- Target programmes and supervisors ----

@dataclass
class PhDTarget:
    programme: str
    institution: str
    country: str
    deadline: str          # YYYY-MM-DD
    funding: str           # 'fully-funded', 'competitive-fellowship', 'fee-waiver-only'
    theme: str
    potential_supervisors: List[str]
    recent_papers: List[str]  # at least 2 to read
    notes: str = ''
    application_started: bool = False

TARGET_PROGRAMMES: List[PhDTarget] = [
    PhDTarget(
        programme='EMBL International PhD Programme (EIPP)',
        institution='EMBL (multiple campuses)',
        country='Germany/UK/Spain/France',
        deadline='2026-09-15',
        funding='fully-funded',
        theme='Bioinformatics, genomics, structural biology, AI for biology',
        potential_supervisors=[
            'Jan Korbel (structural genomics, cancer)',
            'Judith Zaugg (regulatory genomics, chromatin)',
            'Tobias Rausch (long-read genomics)',
        ],
        recent_papers=[
            'Korbel & Campbell (2013) Nature Reviews Genetics — structural variants',
            'Zaugg et al. (2022) Nature Genetics — TF binding and chromatin',
        ],
        notes='Two-round process: online form (Sep), then on-site interview (Jan). Must match with a group BEFORE submitting.',
    ),
    PhDTarget(
        programme='IMPRS for Molecular Life Sciences',
        institution='Max Planck / LMU Munich',
        country='Germany',
        deadline='2026-11-15',
        funding='fully-funded',
        theme='Computational genomics, systems biology, quantitative biology',
        potential_supervisors=[
            'Fabian Theis (single-cell ML, CellRank)',
            'Niklas Doehner (regulatory networks)',
        ],
        recent_papers=[
            'Bergen et al. (2020) Nature Biotechnology — RNA velocity scVelo',
            'Lotfollahi et al. (2023) Nature Methods — scGen perturbation modeling',
        ],
        notes='Fabian Theis group is one of the most directly relevant for CP02 (sequence models, single-cell).',
    ),
    PhDTarget(
        programme='Wellcome Sanger Institute PhD Programme',
        institution='Wellcome Sanger Institute',
        country='UK',
        deadline='2026-11-01',
        funding='fully-funded',
        theme='Genomics, cancer genomics, malaria, infectious disease',
        potential_supervisors=[
            'Leopold Parts (population genomics, ML for biology)',
            'Sarah Teichmann (single-cell, spatial transcriptomics)',
        ],
        recent_papers=[
            'Parts et al. (2021) Nature Genetics — genetic effects on gene expression',
            'Teichmann et al. (2022) Nature — spatial transcriptomics atlas',
        ],
        notes='Strong Track A connection: Parts group works on deep learning + genomics, directly relevant to CP02.',
    ),
    PhDTarget(
        programme='Helmholtz Munich / HMGU Graduate School',
        institution='Helmholtz Munich',
        country='Germany',
        deadline='2026-12-01',
        funding='fully-funded',
        theme='Computational health, diabetes/metabolism, AI for precision medicine',
        potential_supervisors=[
            'Fabian Theis (ML for single-cell — same as IMPRS)',
            'Carsten Marr (image analysis, cell classification)',
        ],
        recent_papers=[
            'Dann et al. (2022) Nature Methods — differential abundance testing',
        ],
        notes='Good fallback; deadline is latest of the four. Theis group jointly affiliated IMPRS/HMGU.',
    ),
]

print('=== Target PhD Programmes ===')
for p in TARGET_PROGRAMMES:
    started = '✓' if p.application_started else '○'
    print(f'{started} {p.programme}')
    print(f'   Deadline: {p.deadline} | Funding: {p.funding}')
    print(f'   Theme: {p.theme}')
    print(f'   Supervisors: {" | ".join(p.potential_supervisors[:2])}')
    print()
```

```python
# ---- Application package audit ----

@dataclass
class ApplicationPackage:
    documents: dict  # name -> (done: bool, notes: str)

package = ApplicationPackage(documents={
    'Personal statement (draft 1)':      (False, 'Use PS_TEMPLATE above; target 900-1000 words'),
    'CV (academic format)':              (False, 'Education, research experience, skills, publications/preprints, software'),
    'Writing sample':                    (False, 'One of the capstone writeups, polished to publication standard'),
    'Research proposal (300-500 words)': (False, 'For EMBL and IMPRS; not all programmes require this'),
    'Reference letters (3)':             (False, 'Request at least 6 weeks before earliest deadline'),
    'ORCID created':                     (False, 'orcid.org — takes 5 minutes; needed for all EU applications'),
    'Zenodo DOI for repo':               (False, 'github.com → zenodo integration; click-through in 10 minutes'),
    'Transcript (official)':             (False, 'Official certified copy; allow 2-4 weeks for institution to issue'),
    'English proficiency (IELTS/TOEFL)': (False, 'EMBL requires IELTS ≥ 7.0 or equivalent; check each programme'),
    'Portfolio README polished':         (False, 'portfolio/ folder linked from root README'),
})

def application_progress(pkg: ApplicationPackage) -> float:
    done = sum(1 for v in pkg.documents.values() if v[0])
    return done / len(pkg.documents)

print('=== PhD Application Package Audit ===')
for doc, (done, notes) in package.documents.items():
    status = '✓' if done else '○'
    print(f'  {status} {doc}')
    if not done:
        print(f'      → {notes}')

print(f'\nOverall progress: {application_progress(package):.0%} ({sum(1 for v in package.documents.values() if v[0])}/{len(package.documents)} items done)')

# Earliest deadline
deadlines = [(p.deadline, p.programme) for p in TARGET_PROGRAMMES]
deadlines.sort()
print(f'\nEarliest deadline: {deadlines[0][0]} ({deadlines[0][1]})')
print(f'Days until earliest deadline (from 2026-07-14): ',
       (np.datetime64(deadlines[0][0]) - np.datetime64('2026-07-14')).astype(int), 'days')
```

```python
# ---- Gantt: application timeline ----
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Panel A: Deadline timeline
ax = axes[0]
programmes  = [p.programme.split('(')[0].strip() for p in TARGET_PROGRAMMES]
deadline_dt = [np.datetime64(p.deadline) for p in TARGET_PROGRAMMES]
start_dt    = np.datetime64('2026-07-14')
days_from_now = [(d - start_dt).astype(int) for d in deadline_dt]
funding_colors = {'fully-funded': '#59a14f', 'competitive-fellowship': '#f28e2b', 'fee-waiver-only': '#e15759'}
colors_d = [funding_colors.get(p.funding, 'grey') for p in TARGET_PROGRAMMES]
bars_d = ax.barh(programmes, days_from_now, color=colors_d, edgecolor='black', alpha=0.8)
for bar, days, prog in zip(bars_d, days_from_now, programmes):
    ax.text(days + 3, bar.get_y() + bar.get_height()/2,
             f'{days}d', va='center', fontsize=9)
ax.set_xlabel('Days from today (2026-07-14) until deadline')
ax.set_title('A. Application deadlines\n(green = fully funded)')
patch_ff = mpatches.Patch(color='#59a14f', alpha=0.8, label='Fully funded')
ax.legend(handles=[patch_ff], fontsize=8)
ax.text(-0.12, 1.05, 'A', transform=ax.transAxes, fontsize=12, fontweight='bold')

# Panel B: Action items by priority
ax = axes[1]
ax.axis('off')
action_items = [
    ('NOW', '#e15759',  [
        'Create ORCID (10 min)',
        'Create Zenodo DOI (10 min)',
        'Email referees: earliest deadline is Sep 15',
        'Request official transcript from institution',
    ]),
    ('MONTH 1 (Aug 2026)', '#f28e2b', [
        'Write personal statement draft 1 (use PS_TEMPLATE)',
        'Polish portfolio README + portfolio/ folder',
        'Read 2 papers from each target supervisor',
        'Check IELTS/TOEFL requirement per programme',
    ]),
    ('MONTH 2 (Sep 2026)', '#4e79a7', [
        'Submit EMBL application (deadline Sep 15)',
        'Finalise CV with quantified capstone results',
        'Complete Sanger application (deadline Nov 1)',
    ]),
    ('MONTH 3+ (Oct–Dec 2026)', '#59a14f', [
        'Submit IMPRS (Nov 15) and HMGU (Dec 1)',
        'Prepare for interview (Cold-paper explanation, Track A drills)',
        'Continue paper reading log in paper-notes/',
    ]),
]
y_start = 0.97
for period, color, items in action_items:
    ax.text(0.0, y_start, period, transform=ax.transAxes, fontsize=10,
             fontweight='bold', color=color, va='top')
    y_start -= 0.04
    for item in items:
        ax.text(0.03, y_start, f'• {item}', transform=ax.transAxes, fontsize=8, va='top')
        y_start -= 0.04
    y_start -= 0.02
ax.set_title('B. Prioritized action items')
ax.text(-0.08, 1.05, 'B', transform=ax.transAxes, fontsize=12, fontweight='bold')

plt.suptitle('Module 20: PhD Application Package', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('phd_application.png', dpi=150, bbox_inches='tight')
plt.show()
```
