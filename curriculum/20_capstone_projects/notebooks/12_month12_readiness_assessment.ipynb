```python
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass, field
from typing import List, Dict, Optional

# ---- 50-criterion readiness scorecard ----

@dataclass
class Criterion:
    text: str
    track: str   # 'A', 'B', 'both'
    domain: str  # 'method', 'code', 'communication', 'interview', 'application'
    weight: int  # 1=standard, 2=important, 3=critical
    done: bool = False

SCORECARD: List[Criterion] = [
    # ---- TRACK A: Methods ----
    Criterion('Can explain BLAST (seed-and-extend, scoring matrix, E-value) without notes', 'A', 'interview', 3, False),
    Criterion('Can explain the steps in an NGS QC pipeline (FastQC → trimming → alignment → QC)', 'A', 'interview', 3, False),
    Criterion('Can explain RNA-seq differential expression end-to-end (normalization → test → FDR)', 'A', 'interview', 3, False),
    Criterion('Knows BWA-MEM vs STAR vs HISAT2: when to use which', 'A', 'method', 2, False),
    Criterion('Can explain SAM/BAM format: FLAG field, CIGAR string, MAPQ', 'A', 'interview', 2, False),
    Criterion('Can explain GATK HaplotypeCaller workflow at a high level', 'A', 'method', 2, False),
    Criterion('Understands p-value vs FDR: can describe BH correction in 2 sentences', 'A', 'interview', 3, False),
    Criterion('Can run a t-test, Mann-Whitney U, and Wilcoxon in Python without docs', 'A', 'code', 2, False),
    Criterion('Knows the difference between AUROC and AUPRC; knows when AUPRC is preferred', 'A', 'interview', 2, False),
    Criterion('Can explain k-fold cross-validation; knows what data leakage is', 'A', 'interview', 3, False),
    Criterion('Can explain what a confusion matrix is and derive precision/recall from it', 'A', 'interview', 2, False),
    Criterion('Understands pairwise sequence alignment: Smith-Waterman vs Needleman-Wunsch', 'A', 'method', 2, False),
    Criterion('Can explain FASTA/FASTQ format; knows Phred quality scores', 'A', 'interview', 2, False),
    Criterion('Comfortable with Unix: grep, awk, sort, uniq, cut, pipe on a FASTQ file', 'A', 'code', 3, False),
    Criterion('Can explain what a VCF file contains (CHROM, POS, REF, ALT, FILTER, INFO, FORMAT)', 'A', 'method', 2, False),
    # ---- TRACK A: Code ----
    Criterion('Can write a working Python script from scratch in ≤ 30 min for a bioinformatics task', 'A', 'code', 3, False),
    Criterion('Public GitHub portfolio with ≥ 3 polished, runnable projects', 'A', 'code', 3, False),
    Criterion('Has at least one project with a README, tests, and reproducible setup', 'A', 'code', 2, False),
    Criterion('Comfortable with Pandas: merge, groupby, pivot, apply on a genomics dataframe', 'A', 'code', 2, False),
    Criterion('Comfortable with Biopython: parse FASTA/FASTQ, run BLAST from Python', 'A', 'code', 2, False),
    # ---- TRACK A: Communication ----
    Criterion('Can explain any CP (01, 02, 03) in 2 minutes to a non-specialist', 'A', 'communication', 3, False),
    Criterion('Can explain the biological motivation of each capstone (not just the algorithm)', 'A', 'interview', 3, False),
    Criterion('Has sent ≥ 3 Track A applications (RA/JRF/Bioinformatics Engineer)', 'A', 'application', 3, False),
    Criterion('Has received ≥ 1 interview invitation for a Track A position', 'A', 'application', 2, False),
    Criterion('Can describe your strongest original contribution in 1 sentence', 'A', 'communication', 2, False),
    # ---- TRACK B: Methods ----
    Criterion('Has read ≥ 5 papers with Pass-3 reconstruction notes in paper-notes/', 'B', 'method', 3, False),
    Criterion('Can cold-explain an unseen bioinformatics abstract in ≤ 5 minutes', 'B', 'interview', 3, False),
    Criterion('Understands CNNs for sequence: Conv1d, pooling, receptive field, motif recovery', 'B', 'method', 3, False),
    Criterion('Understands GNNs: message passing, GCN, GAT, PPI network application', 'B', 'method', 2, False),
    Criterion('Understands MCMC: Metropolis-Hastings, burn-in, mixing, trace diagnostics', 'B', 'method', 3, False),
    Criterion('Understands likelihood-ratio tests and AIC model comparison', 'B', 'method', 2, False),
    Criterion('Understands single-cell RNA-seq: UMI, doublets, clustering, UMAP, trajectory', 'B', 'method', 2, False),
    Criterion('Understands AlphaFold2 at a conceptual level: MSA, pairwise representation, confidence', 'B', 'method', 2, False),
    Criterion('Can derive the SIR ODE from first principles in ≤ 5 minutes', 'B', 'method', 2, False),
    Criterion('Knows what R₀ means and how to estimate it from an epidemic curve', 'B', 'interview', 2, False),
    # ---- TRACK B: Application ----
    Criterion('Has ORCID registered', 'B', 'application', 2, False),
    Criterion('Has Zenodo DOI for the repository', 'B', 'application', 2, False),
    Criterion('Personal statement draft 1 written (≥ 900 words)', 'B', 'application', 3, False),
    Criterion('Has read ≥ 2 papers from each target supervisor', 'B', 'application', 3, False),
    Criterion('Reference letters requested ≥ 6 weeks before earliest deadline', 'B', 'application', 3, False),
    Criterion('Official transcript obtained', 'B', 'application', 2, False),
    Criterion('English proficiency documentation obtained (IELTS ≥ 7.0 or equivalent)', 'B', 'application', 2, False),
    Criterion('Has submitted ≥ 1 Track B application', 'B', 'application', 3, False),
    # ---- BOTH: Cross-cutting ----
    Criterion('Repo is fully reproducible from a clean clone (tested)', 'both', 'code', 3, False),
    Criterion('CI/CD pipeline is green on main branch', 'both', 'code', 2, False),
    Criterion('Understands Agile/Git workflow: branch, PR, review cycle', 'both', 'code', 2, False),
    Criterion('Can write a publication-quality figure in matplotlib from scratch', 'both', 'communication', 2, False),
    Criterion('Has completed all 3 capstone projects end-to-end', 'both', 'method', 3, False),
    Criterion('Has given ≥ 1 journal club-style presentation (real or practice)', 'both', 'communication', 2, False),
    Criterion('Weekly progress log maintained for ≥ 6 consecutive months', 'both', 'code', 2, False),
    Criterion('Knows the names of ≥ 3 key groups worldwide in your target sub-field', 'both', 'interview', 2, False),
]

print(f'Total criteria: {len(SCORECARD)}')

def compute_scorecard(scorecard: List[Criterion]) -> Dict:
    for track in ['A', 'B', 'both']:
        items = [c for c in scorecard if c.track in [track, 'both'] or (track == 'both' and c.track == 'both')]
        if items:
            done_w = sum(c.weight for c in items if c.done)
            total_w = sum(c.weight for c in items)
            print(f'Track {track}: {sum(1 for c in items if c.done)}/{len(items)} criteria done ({done_w}/{total_w} weighted)')
    overall_w = sum(c.weight for c in scorecard if c.done)
    total_w   = sum(c.weight for c in scorecard)
    print(f'Overall: {sum(1 for c in scorecard if c.done)}/{len(scorecard)} ({overall_w/total_w:.0%} weighted)')
    return {'overall': overall_w / total_w}

print('\n=== Month-12 Readiness Scorecard ===')
_ = compute_scorecard(SCORECARD)
```

def bh_fdr_threshold(pvalues, alpha=0.05):

```python
# ---- Final scorecard visualization ----
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Panel A: Scorecard by domain
domains = ['method', 'code', 'communication', 'interview', 'application']
domain_colors = {'method': '#4e79a7', 'code': '#59a14f', 'communication': '#f28e2b',
                  'interview': '#e15759', 'application': '#76b7b2'}
domain_scores = {}
for d in domains:
    items = [c for c in SCORECARD if c.domain == d]
    done_w  = sum(c.weight for c in items if c.done)
    total_w = sum(c.weight for c in items)
    domain_scores[d] = done_w / total_w if total_w > 0 else 0.0

ax = axes[0]
bars_d = ax.bar(domains, [domain_scores[d] for d in domains],
                  color=[domain_colors[d] for d in domains], edgecolor='black', alpha=0.8)
for bar, d in zip(bars_d, domains):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{domain_scores[d]:.0%}', ha='center', fontsize=9)
ax.axhline(0.8, color='grey', ls='--', lw=1, label='80% bar')
ax.set_ylim(0, 1.1)
ax.set_ylabel('Weighted completion score')
ax.set_title('A. Month-12 Readiness by Domain\n(80% = ready to apply)')
ax.legend(fontsize=8)
ax.text(-0.12, 1.05, 'A', transform=ax.transAxes, fontsize=12, fontweight='bold')

# Panel B: Track A vs Track B radar
ax = axes[1]
ax.axis('off')

# Radar chart using polar axes inset
import matplotlib
ax_radar = fig.add_axes([0.55, 0.1, 0.4, 0.8], polar=True)
cats_radar = ['Methods', 'Code', 'Communication', 'Interview', 'Application']
N_cats = len(cats_radar)
angles = np.linspace(0, 2*np.pi, N_cats, endpoint=False).tolist()
angles += angles[:1]  # close the polygon

track_A_items = [c for c in SCORECARD if c.track in ('A', 'both')]
track_B_items = [c for c in SCORECARD if c.track in ('B', 'both')]

def track_scores_by_domain(items, domains):
    scores = []
    for d in domains:
        sub = [c for c in items if c.domain == d]
        done_w  = sum(c.weight for c in sub if c.done)
        total_w = sum(c.weight for c in sub)
        scores.append(done_w / total_w if total_w > 0 else 0.0)
    return scores

domain_map = {'Methods': 'method', 'Code': 'code', 'Communication': 'communication',
               'Interview': 'interview', 'Application': 'application'}

scores_A = track_scores_by_domain(track_A_items, [domain_map[c] for c in cats_radar]) + \
           [track_scores_by_domain(track_A_items, [domain_map[c] for c in cats_radar])[0]]
scores_B = track_scores_by_domain(track_B_items, [domain_map[c] for c in cats_radar]) + \
           [track_scores_by_domain(track_B_items, [domain_map[c] for c in cats_radar])[0]]

ax_radar.plot(angles, scores_A, color='#4e79a7', lw=2, label='Track A')
ax_radar.fill(angles, scores_A, color='#4e79a7', alpha=0.15)
ax_radar.plot(angles, scores_B, color='#e15759', lw=2, label='Track B')
ax_radar.fill(angles, scores_B, color='#e15759', alpha=0.15)
ax_radar.set_xticks(angles[:-1]); ax_radar.set_xticklabels(cats_radar, fontsize=8)
ax_radar.set_ylim(0, 1)
ax_radar.set_yticks([0.25, 0.5, 0.75, 1.0])
ax_radar.set_yticklabels(['25%', '50%', '75%', '100%'], fontsize=7)
ax_radar.set_title('B. Track A vs Track B readiness', pad=20, fontsize=10)
ax_radar.legend(loc='upper right', bbox_to_anchor=(1.35, 1.1), fontsize=8)

plt.suptitle('Module 20: Month-12 Readiness Assessment', fontsize=12, fontweight='bold', x=0.35)
plt.savefig('month12_readiness.png', dpi=150, bbox_inches='tight')
plt.show()
print('\nModule 20 complete. The 12-month program is now fully scaffolded.')
```
