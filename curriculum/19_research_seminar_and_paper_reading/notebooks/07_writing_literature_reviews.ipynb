## Motivation

A literature review is not a list of papers. It is a structured argument about the state of a field: what is known, what methods exist, where they fail, and what remains open. Writing one forces you to synthesize rather than summarize — a fundamentally different cognitive task. A strong literature review in a PhD application or a Track A portfolio signals that you understand a field, not just that you have read about it.

## 1. Types of Literature Review

| Type | Scope | Method | When appropriate |
|------|-------|--------|------------------|
| **Narrative review** | Flexible, expert-driven | Author selects papers based on judgment | Emerging field, pedagogical introduction, perspective piece |
| **Systematic review** | Exhaustive, protocol-driven | Pre-specified search, inclusion criteria, PRISMA flow | Clinical/epidemiological questions; when completeness matters |
| **Scoping review** | Broad, mapping-focused | Identify volume and type of evidence, not quality | New field, before a systematic review, identifying gaps |

For PhD applications in computational biology: **narrative review** is most common. You are demonstrating expertise, not performing a clinical synthesis.

## 2. Structure of a Narrative Literature Review

1. **Scope definition** — what is and is not included (by topic, organism, method, date range)
2. **Search strategy** — which databases, which search terms, how many papers found, how many included
3. **Synthesis** — group papers by theme, method, or finding; write argument paragraphs, not paper-by-paper summaries
4. **Gap identification** — what remains unknown or unsolved (this is the most valuable section)
5. **Open questions** — specific, answerable research questions that emerge from the gaps

## 3. Synthesis vs Summary

**Summary (wrong approach):**  
> Smith et al. (2020) found X. Jones et al. (2021) found Y. Zhang et al. (2022) found Z.

**Synthesis (correct approach):**  
> Three independent studies have shown that [general pattern A] (Smith et al. 2020; Jones et al. 2021; Zhang et al. 2022), though the mechanism remains contested: Smith et al. attribute this to [mechanism 1], while Jones et al. propose [mechanism 2].

```python
from dataclasses import dataclass, field
from typing import List, Dict, Optional
from collections import defaultdict, Counter
import numpy as np
import matplotlib.pyplot as plt


@dataclass
class LiteraturePaper:
    title: str
    authors: str
    year: int
    topic: str           # e.g. 'sequence_alignment', 'single_cell', 'protein_structure'
    method: str          # e.g. 'dynamic_programming', 'deep_learning', 'HMM'
    organism: str        # e.g. 'human', 'mouse', 'E. coli', 'Arabidopsis'
    dataset_size: int    # approximate number of samples/cells/sequences
    n_citations: int = 0


class LiteratureMap:
    """Manage a collection of papers and compute field-level statistics."""

    def __init__(self, topic_name: str = 'Computational Biology'):
        self.topic_name = topic_name
        self.papers: List[LiteraturePaper] = []

    def add(self, paper: LiteraturePaper) -> None:
        self.papers.append(paper)

    def method_frequency_over_time(self) -> Dict[str, Dict[int, int]]:
        """Count papers per method per year."""
        freq: Dict[str, Dict[int, int]] = defaultdict(lambda: defaultdict(int))
        for p in self.papers:
            freq[p.method][p.year] += 1
        return {method: dict(years) for method, years in freq.items()}

    def topic_method_matrix(self) -> tuple:
        """Return (topics, methods, count_matrix) for a co-occurrence heatmap."""
        topics = sorted(set(p.topic for p in self.papers))
        methods = sorted(set(p.method for p in self.papers))
        matrix = np.zeros((len(topics), len(methods)))
        for p in self.papers:
            i = topics.index(p.topic)
            j = methods.index(p.method)
            matrix[i, j] += 1
        return topics, methods, matrix

    def identify_gaps(self) -> List[str]:
        """Identify (topic, organism) combinations with < 2 papers."""
        combo_count: Counter = Counter()
        for p in self.papers:
            combo_count[(p.topic, p.organism)] += 1

        all_topics = set(p.topic for p in self.papers)
        all_organisms = set(p.organism for p in self.papers)
        gaps = []
        for topic in all_topics:
            for organism in all_organisms:
                if combo_count[(topic, organism)] < 2:
                    gaps.append(f'{topic} in {organism} ({combo_count[(topic, organism)]} paper(s))')
        return sorted(gaps)


# Populate with 20 synthetic papers across 5 sub-fields
lmap = LiteratureMap('Computational Transcriptomics')

synthetic_papers = [
    # Sequence alignment
    LiteraturePaper('Global alignment of protein sequences', 'Author A', 1970, 'sequence_alignment', 'dynamic_programming', 'human', 100, 5000),
    LiteraturePaper('Local alignment with affine gaps', 'Author B', 1981, 'sequence_alignment', 'dynamic_programming', 'E. coli', 500, 3000),
    LiteraturePaper('BLAST heuristic alignment', 'Author C', 1990, 'sequence_alignment', 'heuristic', 'human', 10000, 80000),
    LiteraturePaper('Ultra-fast read alignment', 'Author D', 2013, 'sequence_alignment', 'FM_index', 'human', 1000000, 8000),
    # Single-cell
    LiteraturePaper('Droplet-based scRNA-seq', 'Author E', 2015, 'single_cell', 'clustering', 'human', 2000, 4000),
    LiteraturePaper('Single-cell integration methods', 'Author F', 2018, 'single_cell', 'deep_learning', 'mouse', 50000, 3000),
    LiteraturePaper('Harmony fast integration', 'Author G', 2019, 'single_cell', 'optimization', 'human', 100000, 2500),
    LiteraturePaper('RNA velocity in single cells', 'Author H', 2018, 'single_cell', 'differential_equations', 'mouse', 30000, 2000),
    LiteraturePaper('scATAC-seq chromatin accessibility', 'Author I', 2019, 'single_cell', 'matrix_factorization', 'human', 80000, 1500),
    # Differential expression
    LiteraturePaper('Negative binomial model for RNA-seq', 'Author J', 2010, 'differential_expression', 'statistical_test', 'human', 5000, 4000),
    LiteraturePaper('DESeq2 moderated fold change', 'Author K', 2014, 'differential_expression', 'statistical_test', 'mouse', 10000, 15000),
    LiteraturePaper('edgeR for count data', 'Author L', 2010, 'differential_expression', 'statistical_test', 'Arabidopsis', 8000, 12000),
    LiteraturePaper('scRNA-seq differential abundance', 'Author M', 2020, 'differential_expression', 'statistical_test', 'human', 50000, 800),
    # Protein structure
    LiteraturePaper('Protein fold recognition by HMM', 'Author N', 1998, 'protein_structure', 'HMM', 'human', 200, 2000),
    LiteraturePaper('Deep learning for protein structure', 'Author O', 2019, 'protein_structure', 'deep_learning', 'human', 10000, 1000),
    LiteraturePaper('AlphaFold2 structure prediction', 'Author P', 2021, 'protein_structure', 'deep_learning', 'human', 350000, 6000),
    LiteraturePaper('Protein language models', 'Author Q', 2022, 'protein_structure', 'deep_learning', 'E. coli', 250000000, 500),
    # Gene regulation
    LiteraturePaper('Transcription factor binding motifs', 'Author R', 2005, 'gene_regulation', 'HMM', 'Arabidopsis', 1000, 1500),
    LiteraturePaper('Chromatin state prediction', 'Author S', 2015, 'gene_regulation', 'deep_learning', 'mouse', 500, 1200),
    LiteraturePaper('Single-cell regulatory networks', 'Author T', 2022, 'gene_regulation', 'matrix_factorization', 'human', 40000, 400),
]

for p in synthetic_papers:
    lmap.add(p)

print(f"LiteratureMap: {len(lmap.papers)} papers across {len(set(p.topic for p in lmap.papers))} topics")
print(f"Methods used: {sorted(set(p.method for p in lmap.papers))}")
gaps = lmap.identify_gaps()
print(f"\nTop 5 coverage gaps:")
for g in gaps[:5]:
    print(f"  - {g}")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Literature Map — Computational Transcriptomics Field Analysis', fontsize=12, fontweight='bold')

# Panel 1: Publication timeline by method
ax1 = axes[0]
method_freq = lmap.method_frequency_over_time()
methods_to_plot = sorted(method_freq.keys())
all_years = sorted(set(p.year for p in lmap.papers))
colors_m = plt.cm.tab10(np.linspace(0, 1, len(methods_to_plot)))

for method, color in zip(methods_to_plot, colors_m):
    year_counts = method_freq[method]
    cumulative = [(y, sum(year_counts.get(yy, 0) for yy in all_years if yy <= y))
                  for y in all_years]
    years_plot, counts_plot = zip(*cumulative)
    ax1.plot(years_plot, counts_plot, marker='o', markersize=4,
             label=method.replace('_', ' '), color=color, linewidth=1.5)

ax1.set_xlabel('Year')
ax1.set_ylabel('Cumulative papers')
ax1.set_title('Method Evolution Over Time')
ax1.legend(fontsize=7, loc='upper left')
ax1.grid(True, alpha=0.3)

# Panel 2: Topic-method co-occurrence heatmap
ax2 = axes[1]
topics, methods, matrix = lmap.topic_method_matrix()
im = ax2.imshow(matrix, cmap='Blues', aspect='auto')
ax2.set_xticks(range(len(methods)))
ax2.set_xticklabels([m.replace('_', '\n') for m in methods], fontsize=7, rotation=45, ha='right')
ax2.set_yticks(range(len(topics)))
ax2.set_yticklabels([t.replace('_', ' ') for t in topics], fontsize=8)
ax2.set_title('Topic × Method Co-occurrence')
plt.colorbar(im, ax=ax2, label='Paper count')
for i in range(len(topics)):
    for j in range(len(methods)):
        if matrix[i, j] > 0:
            ax2.text(j, i, int(matrix[i, j]), ha='center', va='center',
                     fontsize=8, color='white' if matrix[i, j] > 2 else 'black')

# Panel 3: Deep learning rise (method by decade)
ax3 = axes[2]
decades = ['1970s', '1980s', '1990s', '2000s', '2010s', '2020s']
decade_ranges = [(1970, 1979), (1980, 1989), (1990, 1999), (2000, 2009), (2010, 2019), (2020, 2029)]
all_methods_list = sorted(set(p.method for p in lmap.papers))
decade_method_counts = np.zeros((len(all_methods_list), len(decades)))
for p in lmap.papers:
    for di, (d_start, d_end) in enumerate(decade_ranges):
        if d_start <= p.year <= d_end:
            mi = all_methods_list.index(p.method)
            decade_method_counts[mi, di] += 1

bottom = np.zeros(len(decades))
colors_d = plt.cm.Set2(np.linspace(0, 1, len(all_methods_list)))
for mi, (method, color) in enumerate(zip(all_methods_list, colors_d)):
    ax3.bar(decades, decade_method_counts[mi], bottom=bottom,
            label=method.replace('_', ' '), color=color, alpha=0.85)
    bottom += decade_method_counts[mi]

ax3.set_xlabel('Decade')
ax3.set_ylabel('Papers')
ax3.set_title('Method Prevalence by Decade')
ax3.legend(fontsize=7, bbox_to_anchor=(1.02, 1), loc='upper left')
ax3.grid(True, axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('literature_map.png', dpi=150, bbox_inches='tight')
plt.show()
```
