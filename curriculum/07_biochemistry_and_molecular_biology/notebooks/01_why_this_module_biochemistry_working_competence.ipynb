## Step 1 — Motivation

Every computational biology paper operates on data that was generated by biochemistry.
RNA-seq counts come from a biochemical protocol. Mass spec intensities come from
proteolysis and ionisation. ChIP-seq peaks come from antibody-based enrichment. You
cannot critically evaluate what you do not understand at a mechanistic level. This
module closes that gap at the level needed to be useful, not to become a biochemist.

## What Track A interviews actually test from this module

Documented from bioinformatics RA/JRF interview reports (2024–2026 data):

1. **"What does Km mean?"** — expected answer: the substrate concentration at which
   reaction velocity is half of Vmax. A lower Km means higher affinity.
2. **"How does competitive inhibition differ from non-competitive?"** — Km vs. Vmax change.
3. **"Why does RNA-seq data have reads at exon-exon junctions?"** — splicing removes
   introns; junction-spanning reads are the evidence.
4. **"What is a Western blot used for?"** — detecting specific protein by MW and antibody.

These four questions are your assessment targets. Everything else in this module is
context that makes these answers deep rather than recited.
