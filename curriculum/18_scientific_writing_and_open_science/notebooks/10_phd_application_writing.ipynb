## Step 1 — Motivation

EMBL, IMPRS, Wellcome Sanger, OIST, and similar programmes receive 200–1000
applications per cycle. The evaluation committee reads your statement of interest
and research proposal in 3–5 minutes in the first pass. The structure, precision,
and evidence in those documents determine whether you make the shortlist.

---
## Step 2 — What European PhD Programmes Actually Evaluate

Based on publicly available information from EMBL EIPP, IMPRS-LS, Wellcome Sanger,
Helmholtz/HMGU, and similar programmes:

| What they look for | What this means in practice |
|--------------------|--------------------------|
| **Research background** | Evidence of real, completed work (not just coursework) |
| **Critical thinking** | Can you identify what's wrong with current approaches? |
| **Technical competence** | Public code, reproduced results, mathematical fluency |
| **Research fit** | Your interests overlap with at least 2–3 potential supervisors |
| **Writing quality** | Precision, concision, no overclaiming — same standards as a paper |
| **Motivation** | Specific reasons for this programme, not generic enthusiasm |

---
## Step 3 — Personal Statement Structure (1000–1500 words)

| Section | Length | What it contains |
|---------|--------|------------------|
| **P1: Hook + research question** | 100 words | What scientific question drives you; why it matters |
| **P2: Technical background** | 200 words | What you can do — with evidence (not just what you've studied) |
| **P3: Key experience** | 300 words | One or two real projects, what you learned, what you can show |
| **P4: Research fit** | 200 words | Why this programme, which supervisors, which projects |
| **P5: Research proposal sketch** | 200 words | A specific question you want to explore (1–3 years) |
| **P6: Closing** | 100 words | What you bring; what you want to take; one sentence on career goal |

**Rules:**
- Name the potential supervisors explicitly. Generic "I would like to work with
  excellent researchers" is a red flag.
- Evidence > assertion. "I am skilled in Python" is an assertion.
  "I built a CNN in PyTorch that achieved AUROC 0.92 on ENCODE ChIP-seq data
  (code at github.com/Vinoth-ai-20/computational-biology)" is evidence.
- Word count discipline: 1000–1500 words. Under-length reads as unprepared;
  over-length reads as unable to prioritize.

---
## Step 4 — Research Proposal Structure (500–1000 words)

Many programmes ask for a separate research proposal. It is structured like the
first 4 paragraphs of a paper introduction, plus a brief plan:

1. **Background** (1–2 paragraphs): the biological question and why it is open
2. **Gap** (1 paragraph): what current methods cannot do
3. **Proposed approach** (1–2 paragraphs): what you would do, with which tools
4. **Expected outcomes** (1 paragraph): what you would find and why it would matter
5. **Feasibility** (1 paragraph): why this is achievable in 3–4 years

**Proposal pitfalls:**
- Proposing a completed study (this was already done, you just don't know it)
- Proposing 10 years of work for a 4-year PhD
- Proposing something with no biological question (a methods paper with no application)
- Matching nothing in the lab you're applying to

## Step 5 — Specific Programme Notes

### EMBL EIPP (European Bioinformatics Institute, Heidelberg / Cambridge)
- Computational focus. Strong interest in: ML/AI + biology, single-cell, structural,
  evolutionary genomics.
- Rotation programme: 3 rotations, then choose a lab. Apply broadly to supervisors.
- Name ≥3 EMBL group leaders whose work you know in detail.

### IMPRS-LS (Munich)
- Max Planck institutes. Molecular life sciences. Computational and systems biology.
- Apply via the unified IMPRS portal. Statement + research proposal required.
- Strong fit: computational neuroscience, systems biology, evolutionary genomics.

### Wellcome Sanger Institute
- UK. Genomics and computational biology heavy.
- Strong fit: NGS, population genetics, cancer genomics, single-cell.

### Helmholtz / HMGU Munich
- Environmental health + computational systems biology. Strong Python/ML culture.
- Strong fit: transcriptomics, epigenomics, network biology, multi-omics.

```python
import re
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# ---- Personal statement auditing tools ----

def audit_personal_statement(text):
    t = text.lower()
    words = len(text.split())
    sentences = [s.strip() for s in re.split(r'[.!?]', text) if len(s.strip()) > 10]

    # Evidence vs assertion
    evidence_signals = ['github', 'auroc', 'accuracy', 'published', 'preprint',
                          'doi', 'poster', 'conference', 'data analysis', 'i built',
                          'i implemented', 'i developed', 'i applied', 'trained']
    assertion_signals = ['i am passionate', 'i am skilled', 'i believe', 'i hope to',
                           'i am eager', 'i would like to', 'in the future']

    evidence_count = sum(1 for e in evidence_signals if e in t)
    assertion_count = sum(1 for a in assertion_signals if a in t)

    # Supervisor named?
    supervisor_named = bool(re.search(r'professor|dr\.|group leader|lab of', t))

    # Specific programme fit?
    programme_specific = bool(re.search(r'embl|imprs|sanger|helmholtz|dkfz|csh|imba', t))

    # Research question present?
    question_present = '?' in text or any(kw in t for kw in
                                             ['how does', 'what determines', 'why do',
                                              'mechanisms', 'role of', 'function of'])

    # Word count in range?
    wc_ok = 900 <= words <= 1600

    return {
        'word_count': words,
        'word_count_ok': wc_ok,
        'evidence_signals': evidence_count,
        'assertion_signals': assertion_count,
        'supervisor_named': supervisor_named,
        'programme_specific': programme_specific,
        'research_question_present': question_present,
        'evidence_ratio': evidence_count / max(evidence_count + assertion_count, 1),
    }

# Two synthetic statements
STATEMENT_WEAK = """
I am passionate about computational biology. I believe that bioinformatics is the future
of medicine. I am skilled in Python and machine learning. I would like to work with
excellent researchers. I hope to contribute to science. I am eager to join your programme.
I believe I would be a good fit. In the future, I want to become a scientist.
"""

STATEMENT_STRONG = """
How does the three-dimensional organization of the genome encode cell-type-specific
gene expression programs? This question drives my research. A CTCF-anchored loop that
contacts an enhancer in one cell type may be absent in another, yet we lack quantitative
models that predict which loops form and what gene expression changes they produce.

Over the past year, I have built a self-supervised curriculum in computational biology
(github.com/Vinoth-ai-20/computational-biology). My work includes a CNN for transcription
factor binding prediction (AUROC 0.92 on 50 ENCODE TFs; Module 14 of the curriculum),
an ODE/network epidemic model comparison (Module 15), and a complete NGS pipeline
for differential expression analysis using DESeq2 (Module 10). All code is reproducible
from a one-command install.

I would like to join the EMBL EIPP programme and work with Dr. Jan Korbel (structural
variation and cancer genomics) or Dr. Wolfgang Huber (statistical genomics methods),
both of whose work on single-cell multi-omics connects directly to my interest in
topologically associating domain dynamics.
"""

for name, stmt in [('WEAK', STATEMENT_WEAK), ('STRONG', STATEMENT_STRONG)]:
    result = audit_personal_statement(stmt)
    print(f'\n=== {name} statement ===')
    for k, v in result.items():
        if isinstance(v, float): print(f'  {k}: {v:.2f}')
        else: print(f'  {k}: {v}')
```

```python
# Visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 7))

# Panel A: Statement audit comparison
ax = axes[0]
criteria = ['Word count\n(900–1600)', 'Evidence\nsignals', 'Assertion\nsignals (low)', 'Supervisor\nnamed', 'Programme\nspecific', 'Research Q\npresent']

weak_r   = audit_personal_statement(STATEMENT_WEAK)
strong_r = audit_personal_statement(STATEMENT_STRONG)

def score_statement(r):
    return [
        1.0 if r['word_count_ok'] else 0.0,
        min(1.0, r['evidence_signals'] / 4),
        1.0 - min(1.0, r['assertion_signals'] / 4),
        1.0 if r['supervisor_named'] else 0.0,
        1.0 if r['programme_specific'] else 0.0,
        1.0 if r['research_question_present'] else 0.0,
    ]

weak_s   = score_statement(weak_r)
strong_s = score_statement(strong_r)

x = np.arange(len(criteria))
w = 0.35
ax.bar(x - w/2, weak_s,   w, label='Weak',   color='tomato',    edgecolor='black', alpha=0.8)
ax.bar(x + w/2, strong_s, w, label='Strong', color='steelblue', edgecolor='black', alpha=0.8)
ax.set_xticks(x); ax.set_xticklabels(criteria, fontsize=8)
ax.set_ylim(0, 1.2); ax.set_ylabel('Score (1.0 = meets criterion)')
ax.set_title('A. Personal statement audit')
ax.legend(fontsize=9)
ax.axhline(1.0, color='grey', ls=':', lw=0.8)

# Panel B: Target programmes
ax = axes[1]
ax.axis('off')
programmes = [
    ('EMBL EIPP', 'Heidelberg/Cambridge', 'ML+bio, single-cell, structural,\nevolutionary genomics', '#4e79a7'),
    ('IMPRS-LS', 'Munich (Max Planck)', 'Systems bio, comp neuro,\nevolutionary genomics', '#f28e2b'),
    ('Wellcome Sanger', 'Cambridge UK', 'NGS, population genetics,\ncancer genomics, single-cell', '#59a14f'),
    ('Helmholtz/HMGU', 'Munich', 'Transcriptomics, epigenomics,\nnetwork biology, multi-omics', '#e15759'),
]
for i, (prog, location, fit, color) in enumerate(programmes):
    y = 0.92 - i * 0.22
    ax.add_patch(mpatches.FancyBboxPatch((0.02, y-0.1), 0.95, 0.18,
                                           boxstyle='round,pad=0.02', facecolor=color,
                                           alpha=0.25, transform=ax.transAxes))
    ax.text(0.5, y + 0.04, f'{prog} — {location}', transform=ax.transAxes,
              fontsize=9, fontweight='bold', ha='center', va='center')
    ax.text(0.5, y - 0.03, fit, transform=ax.transAxes, fontsize=8,
              ha='center', va='center', style='italic', color='#444')
ax.set_title('B. Target programmes for Track B\n(as of July 2026)')

# Panel C: Evidence vs assertion ratio
ax = axes[2]
categories_ev = ['Evidence signals', 'Assertion signals']
weak_ev   = [weak_r['evidence_signals'],   weak_r['assertion_signals']]
strong_ev = [strong_r['evidence_signals'], strong_r['assertion_signals']]
x = np.arange(2)
ax.bar(x - 0.2, weak_ev,   0.4, label='Weak',   color=['steelblue', 'tomato'], edgecolor='black', alpha=0.8)
ax.bar(x + 0.2, strong_ev, 0.4, label='Strong', color=['steelblue', 'tomato'], edgecolor='black', alpha=0.5)
ax.set_xticks([0, 1]); ax.set_xticklabels(categories_ev)
ax.set_ylabel('Count')
ax.set_title('C. Evidence vs. assertion count\n(evidence should dominate)')
for i, (w_val, s_val) in enumerate(zip(weak_ev, strong_ev)):
    ax.text(i - 0.2, w_val + 0.1, f'Weak: {w_val}', ha='center', fontsize=8)
    ax.text(i + 0.2, s_val + 0.1, f'Strong: {s_val}', ha='center', fontsize=8)

plt.suptitle('Module 18 NB10: PhD Application Writing', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('phd_application_writing.png', dpi=150, bbox_inches='tight')
plt.show()
```
