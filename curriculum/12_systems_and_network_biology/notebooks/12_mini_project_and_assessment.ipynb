def bfs_shortest_path(adj, source, target):
    from collections import deque
    if source == target: return 0
    dist = {source: 0}
    q = deque([source])
    while q:
        u = q.popleft()
        for v in adj[u]:
            if v not in dist:
                dist[v] = dist[u] + 1
                if v == target: return dist[v]
                q.append(v)
    return -1

def local_clustering_coefficient(adj, v):
    neighbors = list(adj[v])
    k = len(neighbors)
    if k < 2: return 0.0
    triangles = sum(1 for i in range(k) for j in range(i+1,k)
                    if neighbors[j] in adj[neighbors[i]])
    return triangles / (k*(k-1)/2)
