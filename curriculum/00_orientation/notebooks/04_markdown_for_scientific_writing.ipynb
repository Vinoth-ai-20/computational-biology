## Step 1 — Motivation

Every notebook in this program is a document. Documentation is not a chore added
after the code is written — it is the medium through which you demonstrate
understanding. A notebook that runs but cannot be read is not a portfolio artifact.

Markdown is the writing language of this program:
- Notebook markdown cells
- All `README.md`, `references.md`, `papers.md`, `roadmap.md` files
- `paper-notes/` logs
- GitHub issues and PR descriptions
- The `progress/` weekly review entries

Mathematics in notebooks uses KaTeX (a subset of LaTeX) via Jupyter's MathJax renderer.

---
## Step 2 — Intuition

Markdown is **prose with lightweight markup** — a small number of conventions
(a `#` for headers, `**` for bold, backticks for code) that render to clean,
readable HTML without an editor that hides the syntax.

The guiding principle: **write for a stranger reading the rendered output.**
Not for yourself, not as notes — as documentation. A notebook README should
answer "what is this, what does it do, and how do I run it?" for someone
who has never seen it before.

## Step 4 — Mathematical Explanation

Mathematics in Markdown uses LaTeX notation rendered by MathJax / KaTeX.

**Inline math:** `$f(x) = x^2$` renders as $f(x) = x^2$

**Display math (centered, on its own line):**

$$
P(\theta \mid D) = \frac{P(D \mid \theta)\, P(\theta)}{P(D)}
$$

This is Bayes' theorem. You will write expressions like this throughout
Modules 03, 04, 05, 08, 09, and 13. Practice the notation now.

**Useful LaTeX for computational biology:**

| Expression | LaTeX | Renders as |
|-----------|-------|-----------|
| Subscript | `x_{ij}` | $x_{ij}$ |
| Superscript | `e^{-\lambda t}` | $e^{-\lambda t}$ |
| Fraction | `\frac{a}{b}` | $\frac{a}{b}$ |
| Sum | `\sum_{i=1}^{n} x_i` | $\sum_{i=1}^{n} x_i$ |
| Product | `\prod_{i=1}^{n}` | $\prod_{i=1}^{n}$ |
| Hardy-Weinberg | `p^2 + 2pq + q^2 = 1` | $p^2 + 2pq + q^2 = 1$ |
| GC content | `\text{GC} = \frac{G + C}{A + T + G + C}` | $\text{GC} = \frac{G + C}{A + T + G + C}$ |

## Step 6 — Python Implementation

````python
# Cell 6.1 — Render Markdown dynamically in a notebook
from IPython.display import Markdown, display

sample_markdown = """
## GC Content Formula

The GC content of a DNA sequence is defined as:

$$
\\text{GC} = \\frac{G + C}{A + T + G + C}
$$

where $G$, $C$, $A$, $T$ are counts of each nucleotide.

For the sequence `ATGCGCTA`:
- $G = 2$, $C = 2$, $A = 2$, $T = 2$
- $\\text{GC} = \\frac{2+2}{8} = 0.5$

**Key property:** GC content is always in $[0.0, 1.0]$.
"""

display(Markdown(sample_markdown))
````

````python
# Cell 6.2 — Generate a module README template programmatically
def make_module_readme(module_num: str, name: str, tier: str, weeks: str,
                        track_a: str, track_b: str, nb_count: int) -> str:
    return f"""# Module {module_num} — {name}

**Depth tier:** {tier}
**Week budget:** {weeks}
**Notebooks:** {nb_count}
**Status:** ⏳ Not started — 0 / {nb_count} notebooks

---

## Motivation

[One paragraph: why this module exists, what real problem it solves.]

---

## Learning Objectives

By the end of this module you will be able to:

1. [Objective 1]
2. [Objective 2]
3. [Objective 3]

---

## Track A Relevance

**{track_a}**

## Track B Relevance

**{track_b}**

---

## Prerequisites

- [List prerequisite modules or prior knowledge]

---

*See also: [roadmap.md](roadmap.md) — [references.md](references.md) — [papers.md](papers.md)*
"""

# Print a sample for Module 03
sample = make_module_readme(
    module_num="03",
    name="Statistics & Probability",
    tier="Tier 1 — Master",
    weeks="Weeks 2–10 (parallel with Modules 01, 02, 05)",
    track_a="Critical — the single most interview-tested non-coding skill.",
    track_b="Required for reading any paper critically and for all genomics modules.",
    nb_count=18,
)
print(sample)
````
