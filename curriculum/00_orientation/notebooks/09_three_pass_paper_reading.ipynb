## Step 1 — Motivation

Reading a paper is not the same as understanding it. Every researcher knows the feeling of having "read" a paper and being unable to describe it a week later. The Three-Pass method solves this by making understanding — not completion — the goal.

Wilson et al. (2017) is assigned first because it is the checklist this program applies to itself. If you cannot reconstruct its 24 practices from memory, you cannot use it as an audit tool in Module 16 and Assignment A02.

---
## Step 2 — Intuition

Think of reading a paper like reading a map before a hike:

- **Pass 1 (Skim):** Glance at the map — identify the start, end, and rough terrain. 5–10 minutes.
- **Pass 2 (Intent):** Read the map carefully, note the specific trail, elevation changes, landmarks. 45–90 minutes.
- **Pass 3 (Reconstruct):** Draw the map from memory. Where did you have to look it up? Those gaps are what you didn't learn.

The `pass3_unaided` field is the hike, not the map-reading.

## Step 5 — Computational Explanation

**The Three-Pass Method (Keshav, 2007):**

| Pass | Time | Goal | What you read |
|------|------|------|---------------|
| Pass 1 — Skim | 5–10 min | Bird's-eye view | Title, abstract, section headings, figures/captions, conclusion |
| Pass 2 — Intent | 45–90 min | Understand method and evidence | Full paper in order; mark every unclear section before asking |
| Pass 3 — Reconstruct | 1–2 hrs | Ownership of the core idea | Paper closed; re-derive or re-explain the core method from memory |

**What to log in `paper-notes/` for Pass 3:**
- What you could reconstruct unaided (the successes)
- What you had to look up (the gaps — these are the actual learning points)
- `pass3_unaided: true/false/partial` — be honest; partial is the most common starting state

**Track A direct connection:**  
The Track A "explain this paper" interview pattern is Pass 3 under time pressure with an audience.  
Every Pass-3 session in this program is rehearsal for that exact moment.

## Step 6 — Python Implementation

```python
# Cell 6.1 — Open the paper-notes file and print a reminder of the Pass structure
import pathlib

repo_root = pathlib.Path.cwd()
while repo_root != repo_root.parent:
    if (repo_root / ".git").exists():
        break
    repo_root = repo_root.parent

note_path = repo_root / "paper-notes" / "wilson_2017_good_enough_practices.md"
if note_path.exists():
    print(f"Note file found: {note_path.relative_to(repo_root)}")
    print()
    # Print the frontmatter
    content = note_path.read_text(encoding="utf-8")
    frontmatter_end = content.find("---", 3)
    if frontmatter_end > 0:
        print(content[:frontmatter_end + 3])
else:
    print("Note file not found. Run Notebook 09 (Literature Review Workflow) first.")
```

```python
# Cell 6.2 — Pass 1 Timer (run this cell, then skim the paper, then stop)
import time
import datetime

PASS1_BUDGET_SECONDS = 600  # 10 minutes

print("PASS 1 — SKIM")
print("=" * 40)
print(f"Budget: {PASS1_BUDGET_SECONDS // 60} minutes")
print(f"Paper:  Wilson et al. (2017)")
print(f"DOI:    https://doi.org/10.1371/journal.pcbi.1005510")
print()
print("What to read:")
print("  1. Title and abstract (2 min)")
print("  2. All section headings (1 min)")
print("  3. All figures and their captions (3 min)")
print("  4. Conclusion section (2 min)")
print()
print(f"Start time: {datetime.datetime.now().strftime('%H:%M:%S')}")
print("GO. Come back when 10 minutes have elapsed.")
```

```python
# Cell 6.3 — After Pass 1: answer these questions before Pass 2
# Write your answers in the paper-notes file, not here.
# Then run Pass 2 (read fully — budget 60 minutes).

pass1_questions = [
    "What is the paper's main claim in one sentence?",
    "Who is the intended audience?",
    "How many 'good enough practices' does the paper describe?",
    "What is the paper's key distinguishing stance vs. 'best practices'?",
    "Does this paper seem relevant to this program? (Yes/No + why in 10 words)",
]

print("Pass 1 debrief questions (answer before starting Pass 2):")
for i, q in enumerate(pass1_questions, 1):
    print(f"  {i}. {q}")

print()
print("Record your answers in: paper-notes/wilson_2017_good_enough_practices.md")
print("Then proceed to Pass 2 (60 minutes, read the full paper).")
```

```python
# Cell 6.4 — Pass 3: Reconstruct the 24 practices from memory
# Run this cell AFTER completing Pass 2. Paper closed.
# Try to fill in the list below from memory before looking at the paper.

print("PASS 3 — RECONSTRUCT")
print("=" * 40)
print("Paper closed. Fill in as many of the 24 practices as you can from memory.")
print("Use the structure: [Category] — [Practice]")
print()
print("Categories in Wilson et al. (2017):")
categories = [
    "Data management",
    "Software",
    "Collaboration",
    "Project organization",
    "Keeping track of changes",
    "Writing",
]
for c in categories:
    print(f"  - {c}")

print()
print("Write your reconstruction in: paper-notes/wilson_2017_good_enough_practices.md")
print("Then open the paper and check. Mark every gap in the note file.")
print("Set pass3_unaided: true / false / partial based on what you could and could not recall.")
```

## Step 7 — Visualization

```python
# Cell 7.1 — Track Pass-3 scores across papers over time
# This will grow as more papers are read. Start with Wilson et al.

import pathlib

paper_notes_dir = repo_root / "paper-notes"
note_files = sorted(paper_notes_dir.glob("*.md"))
note_files = [f for f in note_files if f.name != "README.md"]

print(f"paper-notes/ summary ({len(note_files)} papers):")
print(f"{'File':<50} {'pass3_unaided'}")
print("-" * 65)

for f in note_files:
    content = f.read_text(encoding="utf-8")
    # Extract pass3_unaided from frontmatter
    p3 = "not recorded"
    for line in content.splitlines():
        if line.startswith("pass3_unaided:"):
            val = line.split(":", 1)[1].strip()
            p3 = val if val else "(blank)"
            break
    print(f"  {f.name:<48} {p3}")
```
