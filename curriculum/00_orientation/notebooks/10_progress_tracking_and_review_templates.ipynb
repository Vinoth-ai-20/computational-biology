This notebook sets up the three review loops (weekly/monthly/quarterly) and creates the first weekly log entry in `progress/`.

---
## Step 1 — Motivation

A 12-month program without a feedback loop is a 12-month wish. The three review loops are the mechanism that turns a plan into a trajectory:

- **Weekly:** are you moving? (three questions; 15 minutes)
- **Monthly:** are the readiness scores trending right? are Track A/B deadlines tracked?
- **Quarterly:** should tier assignments change? is a sub-field pulling ahead?

The `progress/` folder is also evidence. A Track B PhD application that includes a public record of weekly reviews signals sustained commitment — a rare signal.

---
## Step 2 — Intuition

Think of the three loops as three time horizons:

| Loop | Time horizon | Key question |
|------|-------------|-------------|
| Weekly | This week | Did anything move? |
| Monthly | Last 30 days | Is the trajectory right? |
| Quarterly | Last 90 days | Should the plan change? |

The weekly loop catches drift before it becomes a month of lost momentum.
The quarterly loop is the only time tier assignments change — not casually mid-month.

## Step 5 — Computational Explanation

**Weekly review — 3 questions (answer every Sunday, ~15 min):**
1. Which module(s)/notebook(s) moved forward this week, specifically?
2. Any paper completed through Pass 3 — did it succeed unaided?
3. One sentence: what got *finished* (not "worked on" — finished).

**Monthly review — 4 questions (first Sunday of each month, ~45 min):**
1. Recompute readiness scores from `Learning_Progress.md` §13.
2. Track A: applications sent, responses received, interviews scheduled?
3. Track B: any programme deadlines inside the next 60 days?
4. Repository health: commit frequency, any module untouched for 4+ weeks?

**Quarterly checkpoint — explicit re-evaluation (Months 3, 6, 9, 12):**
1. Is a sub-field pulling ahead based on real signal?
2. Should Module 12's tier be promoted (Tier 2 → Tier 1)?
3. Re-baseline the sprint calendar against reality.
4. At Month 9: choose the Capstone (Module 20) topic.

## Step 6 — Python Implementation

```python
# Cell 6.1 — Create the progress/ directory and week 1 log template
import pathlib
import datetime

repo_root = pathlib.Path.cwd()
while repo_root != repo_root.parent:
    if (repo_root / ".git").exists():
        break
    repo_root = repo_root.parent

progress_dir = repo_root / "progress"
progress_dir.mkdir(exist_ok=True)

today = datetime.date.today()
# Week 1 log file name
week1_file = progress_dir / f"{today.year}-W{today.isocalendar()[1]:02d}_weekly.md"

week1_template = f"""# Weekly Review — Week {today.isocalendar()[1]} ({today.strftime('%d %b %Y')})

## 1. What moved this week?

*Which module(s)/notebook(s) specifically moved forward?*

- Module 00 (Orientation): Notebooks 01–11 completed
- [Add other notebooks here]

## 2. Papers through Pass 3

*Any paper completed through Pass 3 this week? Did it succeed unaided?*

- Wilson et al. (2017): pass3_unaided = [true/false/partial]

## 3. What got FINISHED (not worked on — finished)

[One sentence only]

---

## Readiness Score (updated)

*(Update monthly — skip this section for weekly entries unless a milestone was hit)*

---

## Notes / Blockers

[Anything blocking progress? Any unexpected discoveries?]
"""

if not week1_file.exists():
    week1_file.write_text(week1_template, encoding="utf-8")
    print(f"Created: {week1_file.relative_to(repo_root)}")
else:
    print(f"Exists:  {week1_file.relative_to(repo_root)}")

print(f"\nOpen and fill in: {week1_file}")
```

```python
# Cell 6.2 — Generate a monthly review template
month_file = progress_dir / f"{today.strftime('%Y-%m')}_monthly.md"

monthly_template = f"""# Monthly Review — {today.strftime('%B %Y')}

## 1. Readiness Scores (recomputed from Learning_Progress.md §13)

| Scorecard | Previous | This month | Target |
|-----------|----------|------------|--------|
| Research Readiness | — | /10 | Month 6: 4.5/10 |
| Track A (India) | — | /10 | Month 6: 7/10 |
| Track B (Europe PhD) | — | /10 | Month 12: 7.75/10 |

## 2. Track A Status

- Applications sent this month: 0
- Responses received: 0
- Interviews scheduled: 0

## 3. Track B Deadlines (next 60 days)

- Check Learning_Progress.md §7 for programme deadlines

## 4. Repository Health

- Commits this month: [count]
- Modules untouched for 4+ weeks: [list]
- paper-notes/ entries: [count]

## 5. What surprised me this month

[One observation that wasn't in the plan]
"""

if not month_file.exists():
    month_file.write_text(monthly_template, encoding="utf-8")
    print(f"Created: {month_file.relative_to(repo_root)}")
else:
    print(f"Exists:  {month_file.relative_to(repo_root)}")
```

```python
# Cell 6.3 — List existing progress logs
logs = sorted(progress_dir.glob("*.md"))
print(f"progress/ directory contents ({len(logs)} files):")
for f in logs:
    size = f.stat().st_size
    print(f"  {f.name:<45}  {size:>6} bytes")
```
