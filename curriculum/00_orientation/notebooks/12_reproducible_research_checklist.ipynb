## Step 1 — Motivation

A repository that cannot be reproduced from a clean clone is not a reproducible research product — it is a personal archive. The distinction matters:

- A hiring committee that clones your repository and it fails to install is a Track A failure, not just an inconvenience.
- A PhD supervisor who runs the bootstrap and gets a broken environment questions everything else in the repo.

This notebook makes that claim testable: either the one-command bootstrap works, or it does not. "It works on my machine" is not an acceptable answer.

---
## Step 2 — Intuition

Reproducibility has three levels:

| Level | Meaning | What this program requires |
|-------|---------|---------------------------|
| **Repeatable** | The same person, same machine, gets the same result | Minimum — not good enough |
| **Reproducible** | A different person, same machine type, gets the same result | The target for every module notebook |
| **Replicable** | A different person, different approach, gets consistent conclusions | Aspirational — for portfolio artifacts and any publication |

The one-command bootstrap test in this notebook checks the second level.

## Step 5 — Computational Explanation

**The FAIR principles (Wilkinson et al., 2016):**

| Principle | Meaning | How this repository implements it |
|-----------|---------|----------------------------------|
| **F**indable | Data has persistent identifiers | `datasets/README.md` records accession numbers and DOIs |
| **A**ccessible | Data can be retrieved | Raw data fetched by scripts in `scripts/`; fetch scripts are committed |
| **I**nteroperable | Data uses standard formats | FASTA, FASTQ, VCF, CSV — standard biological formats throughout |
| **R**eusable | Data and code are licensed and documented | MIT license; `datasets/README.md` documents each dataset's license |

**The one-command bootstrap:**  
A stranger should be able to clone the repository and set up the environment with a single command. In this program, that command is documented in `ENVIRONMENT.md`. This notebook tests whether that claim is true.

## Step 6 — Python Implementation

````python
# Cell 6.1 — Full Wilson et al. (2017) checklist applied to this repository
import pathlib

repo_root = pathlib.Path.cwd()
while repo_root != repo_root.parent:
    if (repo_root / ".git").exists():
        break
    repo_root = repo_root.parent

# Each tuple: (practice, check_function_or_None, notes)
def check_file_exists(path): return (repo_root / path).exists()
def check_dir_exists(path):  return (repo_root / path).is_dir()

checklist = [
    # DATA MANAGEMENT
    ("Save raw data",                    check_dir_exists("curriculum/01_python_scientific_computing/datasets/raw"),  "datasets/raw/ exists (gitignored)"),
    ("Create analysis-friendly data",    check_dir_exists("curriculum/01_python_scientific_computing/datasets/processed"), "datasets/processed/ exists"),
    ("Record processing steps",          True,                                               "fetch scripts in scripts/"),
    # SOFTWARE
    ("Place code in subdirectory",       check_dir_exists("utilities/compbio_utils"),         "utilities/compbio_utils/"),
    ("README present",                   check_file_exists("README.md"),                     "Root README.md"),
    ("Version control everything",       check_dir_exists(".git"),                           ".git/ directory present"),
    ("License present",                  check_file_exists("LICENSE"),                       "MIT License"),
    # COLLABORATION
    ("ENVIRONMENT.md present",           check_file_exists("ENVIRONMENT.md"),                "setup instructions"),
    ("requirements.txt present",         check_file_exists("requirements.txt"),              "pinned dependencies"),
    # PROJECT ORGANIZATION
    ("Public repository",                True,                                               "github.com/Vinoth-ai-20/computational-biology"),
    ("Raw and processed data separate",  True,                                               "datasets/raw/ vs datasets/processed/"),
    # KEEPING TRACK
    ("Reproducibility toolchain",        check_file_exists(".pre-commit-config.yaml"),       "jupytext + nbstripout pre-commit hooks"),
    ("Tests present",                    check_dir_exists("utilities/compbio_utils/tests"),  "compbio_utils/tests/"),
    # WRITING
    ("Human-readable code",              True,                                               "enforced by Wilson checklist in every notebook"),
    ("Modular code (no giant scripts)",  True,                                               "compbio_utils package architecture"),
]

print(f"Wilson et al. (2017) Checklist — Repository Audit")
print("=" * 65)
passed = 0
for practice, result, notes in checklist:
    status = "✓" if result else "✗"
    if result: passed += 1
    print(f"  {status}  {practice:<40}  {notes}")

print(f"\n{passed}/{len(checklist)} practices met")
if passed < len(checklist):
    print("Fix the ✗ items above before signing off Module 00.")
else:
    print("Repository passes the Wilson et al. audit. Proceed to Module 01.")
````

````python
# Cell 6.2 — Simulate the one-command bootstrap test
# (In practice, this means: clone to a temp directory and run from scratch.)
# Here we verify all the components that the bootstrap relies on.

import sys
import subprocess
import importlib

bootstrap_checks = [
    ("Python ≥ 3.11",           sys.version_info >= (3, 11)),
    ("requirements.txt exists", check_file_exists("requirements.txt")),
    ("ENVIRONMENT.md exists",   check_file_exists("ENVIRONMENT.md")),
    ("numpy importable",        True),  # If this notebook runs, numpy is present
    ("biopython importable",    importlib.util.find_spec("Bio") is not None),
    ("jupytext importable",     importlib.util.find_spec("jupytext") is not None),
    ("nbstripout importable",   importlib.util.find_spec("nbstripout") is not None),
    ("pre-commit config",       check_file_exists(".pre-commit-config.yaml")),
    ("compbio_utils seeded",    check_dir_exists("utilities/compbio_utils")),
    (".git directory",          check_dir_exists(".git")),
]

print("One-Command Bootstrap Verification")
print("=" * 45)
all_pass = True
for name, result in bootstrap_checks:
    status = "✓ PASS" if result else "✗ FAIL"
    if not result: all_pass = False
    print(f"  {status}  {name}")

print()
if all_pass:
    print("All bootstrap checks pass.")
    print("Ready to proceed to Module 01.")
else:
    print("Some checks failed. Resolve before proceeding.")
````

````python
# Cell 6.3 — Module 00 completion gate
# The roadmap.md completion checklist, expressed as code

completion_gate = [
    ("scripts/verify_environment.py passes all checks",          check_file_exists("scripts/verify_environment.py")),
    ("First commit pushed to public repository",                  True),  # Manual check
    ("Jupytext pairing confirmed working",                        check_file_exists(".pre-commit-config.yaml")),
    ("nbstripout confirmed working",                              True),  # Manual check
    ("Pass-3 reconstruction of Wilson et al. logged in paper-notes/", (repo_root / "paper-notes").exists()),
    ("First weekly progress log written in progress/",            any((repo_root / "progress").glob("*weekly*.md")) if (repo_root / "progress").exists() else False),
    ("13-step lesson sequence can be recited unprompted",         True),  # Honor system
    ("One-command bootstrap tested from scratch",                 True),  # Manual test
]

print("Module 00 Completion Gate")
print("=" * 55)
print("Do not start Module 01 until all items below are checked:\n")

gates_passed = 0
for item, passed in completion_gate:
    status = "[✓]" if passed else "[ ]"
    if passed: gates_passed += 1
    print(f"  {status}  {item}")

print(f"\n{gates_passed}/{len(completion_gate)} gate items cleared")
if gates_passed == len(completion_gate):
    print("\n★ Module 00 complete. Update roadmap.md and Learning_Progress.md, then start Module 01.")
else:
    print("\nComplete all items above before starting Module 01.")
````
