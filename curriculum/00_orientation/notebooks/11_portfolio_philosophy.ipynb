---
## Step 1 — Motivation

Not every notebook belongs in `portfolio/`. A rough exploratory notebook is working material; a portfolio artifact is evidence. The distinction matters because a PhD supervisor or hiring manager will spend 3–5 minutes on your repository. They will not read every notebook. They will click `portfolio/` and judge the quality of what they find.

"Defensible cold" is the standard: any artifact in `portfolio/` must be explainable and defensible in a surprise interview setting, without review time.

---
## Step 2 — Intuition

| Working material | Portfolio artifact |
|-----------------|-------------------|
| A notebook that runs | A notebook that runs **top-to-bottom from a clean clone** |
| Code that produces results | Code you can explain **line by line, unprompted** |
| An interesting analysis | An interesting analysis **with a clear question, method, and conclusion** |
| Notes to yourself | Documentation for a stranger |

The polishing checklist (below) is the formal boundary between the two.

## Step 5 — Computational Explanation

**Portfolio artifact polishing checklist:**

- [ ] Runs top-to-bottom from a clean clone (no hidden state, no manual steps)
- [ ] Every function has a one-line docstring
- [ ] Every visualization has a title, axis labels, and units
- [ ] The biological question is stated explicitly at the top
- [ ] The conclusion is stated explicitly at the bottom (not just "results", but what they mean)
- [ ] No magic numbers — all constants are named
- [ ] Dataset source is documented (accession number, license, fetch script)
- [ ] Wilson et al. checklist applied and documented
- [ ] Linked from `portfolio/README.md`
- [ ] Can be explained and defended in a 5-minute cold interview

**Month 3 artifact target:**  
The first portfolio-ready artifact is the Module 08 phylogenetics pipeline — a real, end-to-end sequence analysis on public data. It should be public by end of September 2026.

## Step 6 — Python Implementation

```python
# Cell 6.1 — Create the portfolio/ directory and its README
import pathlib

repo_root = pathlib.Path.cwd()
while repo_root != repo_root.parent:
    if (repo_root / ".git").exists():
        break
    repo_root = repo_root.parent

portfolio_dir = repo_root / "portfolio"
portfolio_dir.mkdir(exist_ok=True)

portfolio_readme = portfolio_dir / "README.md"
if not portfolio_readme.exists():
    portfolio_readme.write_text(
        "# Portfolio\n\n"
        "Polished, defensible artifacts from the 12-month Research Accelerator.\n"
        "Every item here runs top-to-bottom from a clean clone and can be explained\n"
        "and defended in a cold 5-minute interview.\n\n"
        "## Artifacts\n\n"
        "*(Empty — first artifact expected by Month 3, end of September 2026)*\n\n"
        "| Artifact | Module | Status | Track A/B |"
        "\n|---------|--------|--------|----------|",
        encoding="utf-8",
    )
    print(f"Created: {portfolio_readme.relative_to(repo_root)}")
else:
    print(f"Exists:  {portfolio_readme.relative_to(repo_root)}")

print(f"\nportfolio/ is ready. Add the first artifact after Module 08 completes.")
```

```python
# Cell 6.2 — The artifact timeline
planned_artifacts = [
    ("Month 3",  "Module 08", "Phylogenetics pipeline on a real gene family",   "Track A: BLAST/alignment interview; Track B: bioinformatics methods"),
    ("Month 5",  "Module 09", "Full DE analysis on a GEO RNA-seq dataset",      "Track A: Highest posting relevance (RNA-seq); Track B: genomics"),
    ("Month 6",  "Module 12", "Network analysis on a real pathway",              "Track A: Network biology posting (IISER Pune-style); Track B: systems"),
    ("Month 8",  "Module 13", "ML classifier on biological data, full methodology", "Track A/B: ML for biology headline artifact"),
    ("Month 9",  "Module 16", "compbio_utils v1.0 with tests, docs, CI",         "Track A: RSE qualification; Track B: code quality signal"),
    ("Month 12", "Module 20", "Capstone project",                               "Track A/B: Headline of the entire portfolio"),
]

print(f"{'Target':<10} {'Module':<12} {'Artifact':<45} {'Relevance'}")
print("-" * 110)
for month, module, artifact, relevance in planned_artifacts:
    print(f"{month:<10} {module:<12} {artifact:<45} {relevance}")
```
