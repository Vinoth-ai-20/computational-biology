## Step 1 — Motivation

Track A interview pattern #5: "Here is a paper. Explain it back to me."
Track B application: "Demonstrate that you can read primary literature independently."

Both require a practiced workflow, not a one-off effort. The `paper-notes/` folder documents that practice publicly. A PhD supervisor who opens this repository and sees 20 well-structured Pass-3 logs is reading evidence of research maturity, not just enthusiasm.

---
## Step 2 — Intuition

Finding papers is easy. Reading them is the skill.

The difference between "I read that paper" and "I can reconstruct the core method with the paper closed" is the difference between a student and a researcher. This program tracks the second metric, not the first.

Zotero is the reference manager: it stores citation metadata, not PDFs. A full-text PDF of a copyrighted paper never goes into this public repository — only DOIs, metadata, and your personal notes.

## Step 5 — Computational Explanation

**Where to find papers (bookmark all of these):**

| Source | Best for |
|--------|---------|
| [PubMed](https://pubmed.ncbi.nlm.nih.gov/) | Biology/biomedical literature; MeSH search terms |
| [Google Scholar](https://scholar.google.com/) | Fast initial search; citation counts |
| [Semantic Scholar](https://www.semanticscholar.org/) | AI-powered relevance ranking; open-access PDFs |
| [bioRxiv](https://biorxiv.org/) | Biology preprints — newer and often more relevant |
| [arXiv cs.LG / q-bio](https://arxiv.org/) | Computational/ML methods and quantitative biology |
| [EMBL-EBI Training](https://www.ebi.ac.uk/training/) | Bioinformatics tutorials directly relevant to Track A |

The **key field** tracked in every note: `pass3_unaided: true/false/partial`.  
This is the metric. Not "did I read it." Did I reconstruct it.

## Step 6 — Python Implementation

````python
# Cell 6.1 — Scaffold the paper-notes directory
import pathlib

# Find repo root
repo_root = pathlib.Path.cwd()
while repo_root != repo_root.parent:
    if (repo_root / ".git").exists():
        break
    repo_root = repo_root.parent

paper_notes_dir = repo_root / "paper-notes"
paper_notes_dir.mkdir(exist_ok=True)

# Create a README for the paper-notes directory
readme = paper_notes_dir / "README.md"
if not readme.exists():
    readme.write_text(
        "# paper-notes\n\n"
        "One file per paper read anywhere in the program.\n"
        "Each file logs the Three-Pass method output, including whether \n"
        "the Pass-3 reconstruction succeeded unaided.\n\n"
        "**The metric that matters:** `pass3_unaided: true/false/partial`\n\n"
        "Do not commit copyrighted PDF content. Log metadata and personal notes only.\n",
        encoding="utf-8",
    )
    print(f"Created: {readme.relative_to(repo_root)}")
else:
    print(f"Exists:  {readme.relative_to(repo_root)}")

print(f"\npaper-notes/ directory: {paper_notes_dir}")
existing = list(paper_notes_dir.glob("*.md"))
print(f"Existing note files: {len(existing)}")
for f in existing:
    print(f"  {f.name}")
````

````python
# Cell 6.2 — Generate a blank paper-notes template file

def create_paper_note(
    doi: str,
    author_year: str,
    title: str,
    journal: str,
    module: str,
    output_dir: pathlib.Path,
) -> pathlib.Path:
    """Create a blank paper-notes file from the standard template."""
    slug = author_year.lower().replace(" ", "_").replace(",", "")
    filename = output_dir / f"{slug}_{title.split()[0].lower()}.md"

    if filename.exists():
        print(f"Already exists: {filename.name}")
        return filename

    content = f"""---
title: "{title}"
authors: "{author_year}"
journal: "{journal}"
doi: "{doi}"
module: "{module}"
date_read:
pass1_date:
pass2_date:
pass3_date:
pass3_unaided:        # true / false / partial
---

## Pass 1 — Skim (5–10 min)

*Title, abstract, figures, conclusion. What is the one-sentence claim?*

[Write here]

---

## Pass 2 — Intent (45–90 min)

*Read in order. Summarize the method and evidence. Mark every section you don't understand.*

[Write here]

**Fuzzy sections (need re-reading or external explanation):**
- [ ] [Section or concept]

---

## Pass 3 — Reconstruct (1–2 hrs, paper closed)

*Re-explain the core method with the paper closed. What can you reconstruct unaided?*

[Write your reconstruction here]

**What was missing when I tried without the paper:**
[Be specific — this is the learning metric]

---

## Summary

**One-sentence summary (no jargon):**

**Why this paper matters for this program:**

**Connection to other papers/modules:**
"""
    filename.write_text(content, encoding="utf-8")
    return filename


# Create the Wilson et al. (2017) template now — you will fill it in Notebook 10
note_file = create_paper_note(
    doi="10.1371/journal.pcbi.1005510",
    author_year="Wilson et al., 2017",
    title="Good enough practices in scientific computing",
    journal="PLOS Computational Biology",
    module="00_orientation",
    output_dir=paper_notes_dir,
)
print(f"\nNote file: {note_file.relative_to(repo_root)}")
````

````python
# Cell 6.3 — Demonstrate filling the papers.md for Module 01
# The papers.md already exists; this shows what a search session produces.

# A simple PubMed search URL generator
def pubmed_url(query: str) -> str:
    import urllib.parse
    return f"https://pubmed.ncbi.nlm.nih.gov/?term={urllib.parse.quote(query)}"

queries = [
    "BLAST sequence alignment algorithm",
    "DESeq2 differential expression RNA-seq",
    "Hardy-Weinberg equilibrium population genetics",
]

print("Example PubMed search URLs for downstream modules:")
for q in queries:
    print(f"  Query: {q}")
    print(f"  URL:   {pubmed_url(q)}")
    print()
````
