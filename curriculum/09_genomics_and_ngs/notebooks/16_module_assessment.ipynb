# B1 — decode_flag reference
FLAG_BITS = {
    'paired': 0x001, 'proper_pair': 0x002, 'read_unmapped': 0x004,
    'mate_unmapped': 0x008, 'read_reverse': 0x010, 'mate_reverse': 0x020,
    'read1': 0x040, 'read2': 0x080, 'secondary': 0x100,
    'qc_fail': 0x200, 'duplicate': 0x400, 'supplementary': 0x800,
}
def decode_flag(flag): return {n: bool(flag & b) for n, b in FLAG_BITS.items()}

# B2 — compute_tpm reference
def compute_tpm(counts, lengths_bp):
    rate = counts / (lengths_bp / 1000)
    return (rate / rate.sum()) * 1e6

# B3 — genotype_likelihood reference
def genotype_likelihood(bases, quals, genotype):
    a1, a2 = genotype
    log_lik = 0.0
    for b, q in zip(bases, quals):
        eps = 10 ** (-q / 10)
        p = 0.5 * ((1-eps) if b == a1 else eps/3) + 0.5 * ((1-eps) if b == a2 else eps/3)
        log_lik += np.log10(p + 1e-300)
    return log_lik

def call_genotype(bases, quals):
    from itertools import combinations_with_replacement
    genotypes = list(combinations_with_replacement(['A','C','G','T'], 2))
    best = max(genotypes, key=lambda g: genotype_likelihood(bases, quals, g))
    return f'{best[0]}/{best[1]}'
